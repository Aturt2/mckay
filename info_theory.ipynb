```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
trials = np.random.choice(np.arange(1,7), size=10000) + np.random.choice([0,6], size=10000)
```

```python
plt.hist(trials, bins=12)
plt.show()
```

```python
trials_2 = []
for _ in range(1000):
    dice_sum = 0
    for r in range(99):
        dice_sum += np.random.choice(np.arange(6)*6**r)
    trials_2.append(dice_sum)
```

```python
plt.hist(trials2, bins=12)
plt.show()
```

### Exercise 2.17

$q = 1 - p$ (1)<br>
$a = ln \frac{p}{q}$ (2)

(1) $e^a = \frac{p}{q} \rightarrow q = \frac{p}{e^a}$

(2) $\frac{p}{e^a} = 1 - p \rightarrow p(1 + e^a) = e^a \rightarrow p(e^{-a} + 1) = 1$ 

$p = \frac{1}{1+e^{-a}}$

```python
plt.plot(np.arange(-10, 10, 0.1), 1/(1+np.exp(-np.arange(-10, 10, 0.1))))
plt.plot(np.arange(-10, 10, 0.1), np.tanh(np.arange(-10, 10, 0.1)))
plt.legend(["p", "tanh"])
```

```python
plt.plot(np.arange(-10, 10, 0.1), 1/(1+np.exp(-np.arange(-10, 10, 0.1))))
plt.plot(np.arange(-10, 10, 0.1), 1/(1+2**(-np.arange(-10, 10, 0.1))))
plt.legend(["p(a)", "p(b)"])
```

$P(x=0|y) = \frac{P(y|x=0)P(x=0)}{P(y)}$

$\frac{P(x=1|y)}{P(x=0|y)} = \frac{P(y|x=1)P(x=1)}{P(y|x=0)P(x=0)}$

$log\frac{P(x=1|y)}{P(x=0|y)} = log\frac{P(y|x=1)}{P(y|x=0)} + log\frac{P(x=1)}{P(x=0)}$
